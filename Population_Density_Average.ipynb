```python
### this cell is used to import statements for all of the packages used in this project file.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### the csv file will be read at this section
### indexed by country index_col='country'
pop_den = pd.read_csv('population_density_per_square_km.csv' , index_col='country').T # .T is ued to transpose the table
# for easier column manipulation. 
pd.set_option('display.max_rows' , None) # this line is used to display all rows with no split, optinal. 
```

## Data Wrangling

> In this section of the project, data frame will be checked for cleanliness, and then trimed and cleaned for analysis. 
### General Properties

```python
# first of all we need to look for the data description to understand its sturcutre
pop_den.shape
```

```python
pop_den.info() # looking to datafram information
```

```python
pop_den.describe() # describing dataframe and understanding data distribution. 
```

```python
# finding the median 
pop_den_median = np.median(pop_den) # using numpy library
pop_den_median
```

```python
# finding the mean
pop_den_mean = np.mean(pop_den) # using numpy library
pop_den_mean
```

```python
# finding the standard deviation
pop_den_std = np.std(pop_den) # using numpy library
pop_den_std
```

## Exploratory Data Analysis

> This section will compute statistics and create visualizations with the goal of addressing the research questions that are shown below. 

# Research Question 1  (what is the most dense 10 countries in the world?)

# Research Question 2  (what is the least dense 10 countries in the world?)

>

## Part 1  Exploring histograms 


```python
Lower_seg = pop_den.columns[(pop_den_mean < pop_den_median)]
Higher_seg = pop_den.columns[(pop_den_mean > pop_den_median)]
```

```python
# finding histograms of data, in this section list of histograms of countries which mean is less than median. 
pop_den[Lower_seg].hist(figsize=(40,40) , color='green');
```

```python
# finding histograms of data , in this section list of histograms of countries which mean is greater than median. 
pop_den[Higher_seg].hist(figsize=(40,40) , color='blue');
```

> **Comment:** The above historgrams were produced to visualize the frequency of data on every country, this helps understand the distribution 

## Part 2  Exploring Standard Deviations

```python
# this is the list of countris which standard deviation is greater than data mean. 
pop_den[pop_den.columns[(pop_den_std > pop_den_mean)]].hist();
```

>**Comment:*** Found only one country 'Niger'**

#  This section will answer research question 1

```python
# first finding top 10 countries densed in population and assinging it to the variable top10_high
top10_high = pop_den.mean().sort_values(ascending=False).head(10)
```

```python
top10_high # calling a variable for the top 10 highly densed contries in the world
```

```python
# ploting the results
top10_high_plot = top10_high.plot(kind='bar',  title='Most densely populated Countries in the world' 
            , color='red' , figsize=(10,6));
top10_high_plot.set_xlabel(' Countries')
top10_high_plot.set_ylabel(' Population Density Average / km2');
```

> **Comment** As shown in above chart, those countries are filling densely though this doesnt indicate any rise or decline in population. 

#  This section will answer research question 2

```python
# first finding top 10 countries leaest densed in population and assinging it to the variable top10_low
top10_low = pop_den.mean().sort_values(ascending=False).tail(10)
```

```python
top10_low # calling a variable for the top 10 least densed contries in the world
```

```python
# plotting the result 
top10_low_plot = top10_low.plot(kind='bar', title='Least densely populated Countries in the world' 
    , color='green' , figsize=(10,6));
top10_low_plot.set_xlabel(' Countries ')
top10_low_plot.set_ylabel(' Population Density Average / km2');
```

> **Comment** As shown in above chart, those countries are not filling densely though this doesnt indicate any rise or decline in population. 

#  This section will answer research question 3


```python
index_int = pop_den.index.astype(int) # converting string index to int to be used in chart ticks, this has solved the problem of crowded 
```

```python
plt.figure(figsize=(25,10) )
plt.rcParams.update({'font.size': 16})
plt.title('Declining Population Density Average (Samples) ')
plt.xlabel('Years')
plt.ylabel('Population Density Average / km2')
plt.plot(index_int , pop_den['Germany']);
plt.plot(index_int , pop_den['Japan']);
plt.plot(index_int , pop_den['Romania']);
plt.plot(index_int , pop_den['Austria']);
plt.legend()
plt.show()
```

>**Comment:** After testing several countries data, found to be there is countries that population density is declining along the total period, as shown in the above chart. This chart shows sample countries only and more can be retreived, but to avoid congesting the chart, I choosed those ones only.
